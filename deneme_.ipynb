```python
import os
def klasor_yapisi_cikar(dizin, bosluk=""):
    # Gizlenmesini istediДҹin klasГ¶r isimleri
    gizlenecekler = {'.git', '__pycache__', '.venv', 'node_modules', '.idea'}
    
    if os.path.basename(dizin) in gizlenecekler:
        return

    print(f"{bosluk}в”ңв”Җв”Җ {os.path.basename(dizin)}/")
    
    dosyalar = []
    try:
        icerik = os.listdir(dizin)
    except PermissionError:
        return

    for Г¶Дҹe in sorted(icerik):
        yol = os.path.join(dizin, Г¶Дҹe)
        if os.path.isdir(yol):
            klasor_yapisi_cikar(yol, bosluk + "в”Ӯ   ")
        else:
            print(f"{bosluk}в”Ӯ   в””в”Җв”Җ {Г¶Дҹe}")

# Projenin olduДҹu dizini buraya yazabilirsin ('.' Еҹu anki klasГ¶r demek)
proje_yolu = "."
print(f"Proje YapДұsДұ: {os.path.abspath(proje_yolu)}")
klasor_yapisi_cikar(proje_yolu)
```
