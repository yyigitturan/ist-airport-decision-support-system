```python
import pandas as pd

flight_weather_df = pd.read_parquet("/Users/YGT/ist-airport-decision-support-system/data/train/ist_flights_weather.parquet")

scenario_df = pd.read_parquet("/Users/YGT/ist-airport-decision-support-system/data/train/scenario_train_v6.parquet")
```

```python
flight_weather_df.columns
```

```python
scenario_df.columns
```

```python
cols = [
    'scenario_id', 'focusing_flight_id', 'hex_icao', 'entry_time',
    'obs_time', 'scen_idx', 'focusing_emb', 'active_embs', 'prior_embs',
    'active_flight_ids', 'prior_flight_ids', 'n_active', 'n_prior',
    'label_post_terminal_s', 'label_post_terminal_min',
    'airline_name_english', 'callsign_code_iata', 'callsign_code_icao',
    'haul', 'dep_code_iata', 'dep_name_english', 'arr_sched_time_utc',
    'date', 'day_of_week'
]

traj_subset = scenario_df[cols].copy()

merge_keys = [
    'hex_icao',
    'airline_name_english',
    'callsign_code_iata',
    'callsign_code_icao',
    'haul',
    'dep_code_iata',
    'dep_name_english',
    'arr_sched_time_utc',
    'date',
    'day_of_week'

]

full_df = flight_weather_df.merge(
    traj_subset,
    on=merge_keys,
    how='left'
)
```
